```python
import pandas as pd
from scipy import stats

data = pd.read_csv('brain_size.csv', sep=';', quotechar='"', na_values='.')

# Filter data for males
male_data = data[data['Gender'] == 'Male']

# Filter data for females
female_data = data[data['Gender'] == 'Female']

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(male_data['Weight'] - female_data['Weight'], 0)

# Print the results
print('T-Statistic:', t_statistic)
print('P-Value:', p_value)
```

## Simple Linear Regression

```python
import pandas 
import numpy as np
x = np.linspace(-5, 5, 20)
np.random.seed(1)

## normal distributed noise 
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)

## create a data frame containging all the relevant variables 
data = pandas.DataFrame({'x' : x, 'y' : y})

## specify an ordinary leaast squares and fit it 
```

```python
from statsmodels.formula.api import ols 
model = ols ('y ~ x', data). fit()
```

```python
## insepcting the varuous statistics & show regression results
print(model.summary())
```

```python
## Load the data
import os
import pandas 
from urllib.request import urlretrieve

if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages', 'wages.txt')
```

```python
## give names to the columns 
names = [
    'EDUCATION:Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives somewhere else',
    'SEX: 1=Female, 0=Male',
    'EXPERIANCE: Number of years of work experiance', 
    'UNION: 1=Union member, 0=Not a union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Proffesional, 6=Other', 
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried, 1=Married',
]
short_names = [n.split(':') [0] for n in names]
data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None, header=None, engine='python')
data.columns = short_names
```

```python
## Log-transform wages because they are increased in a multiplicative fashion
import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```

```python
## plot various scatter matricies to highlight different aspects of the data
import seaborn
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg')
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg', hue='SEX')
plt.suptitle('Effect of gender: 1=Female, 0=Male')
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'], 
                 kind='reg', hue='RACE')
plt.suptitle('Effect of race:1=other, 2=hispanic, 3-white')
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg', hue='UNION')
plt.suptitle('Effect of union: 1=Union member, 0=Not a Union Member')
```

```python
## Plotting a simple regression
seaborn.lmplot(y='WAGE', x='EDUCATION', data=data)
plt.show()
```

```python
import pandas
import urllib
import os
import statsmodels.formula.api as smf
if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urllib.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')

# EDUCATION: Number of years of education
# SEX: 1=Female, 0=Male
# WAGE: Wage (dollars per hour)
data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None, 
                       header=None, names=['education', 'gender', 'wage'],
                       usecols=[0, 2, 5],
                    )
                       
## convert the genders into strings so that statsmodels can detect that gender is a categorical variable
import numpy as np
data['wage'] = np.log10(data['wage'])
                    
```

```python
import seaborn
seaborn.lmplot(y='wage', x='education', hue='gender', data=data)
```

```python
## statistical regression
import statsmodels.formula.api as sm
```

```python
result = sm.ols(formula='wage ~ education + gender', data=data).fit()
print(result.summary())
```

```python
## there is a different offset in wage and there is a different slope
## model this with an interaction
result=sm.ols(formula='wage ~ education + gender + education * gender', data=data).fit()
print(result.summary())
```

```python
## data does not support hypothesis that education benefits males more than femles 
import matplotlib.pyplot as plt
plt.show()
```
